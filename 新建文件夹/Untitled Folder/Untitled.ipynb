```python
pd.pivot_table(data[['order_id','dishes_name','counts']],index='order_id',columns='dishes_name',values='counts').head()
```

```python
def replace(x):
    import numpy as np
    QU = x.quantile(0.75)
    QL = x.quantile(0.25)
    IQR = QU -QL
    x[(x>(QU+1.5*IQR)) | (x<(QL - 1.5*IQR))] = np.nan
    return x
```
